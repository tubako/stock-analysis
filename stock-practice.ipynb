#### Importing libraries and preparing the environment

```python
import pandas_datareader.data as pdr
import numpy as np
import datetime
import yfinance as yfin
import matplotlib.pyplot as plt
yfin.pdr_override()
%matplotlib inline
plt.style.use('fivethirtyeight')
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
```

#### Reading the stock data and plotting

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2017,1,1)
```

```python
google = yfin.download('GOOGL', start, end)
```

```python
google.head()
```

```python
google['Open'].plot(label= 'GOOGL Open',lw=2,figsize=(15,7))
google['Close'].plot(label= 'GOOGL Close',lw=2)
google['High'].plot(label= 'GOOGL High',lw=2)
google['Low'].plot(label= 'GOOGL Low',lw=2)
plt.legend()
plt.title('Google Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
google['Volume'].plot(figsize=(15,7),lw=2)
plt.title('Volume Traded by Google')
```

### An analysis of several car brands 

Next, various car brands like Tesla, Ford, General Motors will be analyzed to get a general idea of the current car market.

```python
#reading the data
tesla = yfin.download('TSLA', start, end)
ford = yfin.download('F',start,end)
gm = yfin.download('GM',start,end)
```

```python
#downloading the data locally as .csv file
tesla.to_csv('Tesla_Stock.csv')
ford.to_csv('Ford_Stock.csv')
gm.to_csv('GM_Stock.csv') 
```

```python
tesla.head()
```

```python
ford.head()
```

```python
gm.head()
```

```python
tesla['Open'].plot(label='Tesla',figsize=(15,7))
gm['Open'].plot(label='GM')
ford['Open'].plot(label='Ford')

plt.title('Stock Prices for Tesla, GM and Ford')
plt.ylabel('Stock Prices')
plt.legend()
```

```python
#volume traded comparison

tesla['Volume'].plot(label='Tesla',figsize=(15,7))
gm['Volume'].plot(label='GM')
ford['Volume'].plot(label='Ford')

plt.ylabel('Volume Traded')
plt.legend()
```

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
gm['Total Traded'] = gm['Open'] * gm['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
```
